```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('DelayedFlights.csv')
dataset = pd.DataFrame(df)

# grafica 1: una variable categorica
print(dataset.UniqueCarrier)
xpoints = dataset.UniqueCarrier
plt.plot(xpoints, color ='r')
plt.title('UniqueCarrier')
plt.show()
sns.set(style="darkgrid")
sns.boxplot(data=dataset)
plt.title('Retrasos2')
plt.show()

#grafica 2 : Una variable numerica
print(dataset.ArrDelay)
plt.title('ArrDelay')
ypoints = dataset.ArrDelay
plt.plot(ypoints, color ='b')
plt.show()

#grafica 3: Una variable numerica i una categorica

ax1 = dataset.plot.scatter(x='UniqueCarrier', y='ArrDelay', color ='DarkBlue')
plt.title('Retras per companyia')
plt.show()

#grfica 4: Dues variables numriques (ArrDelay i DepDelay)
ax2 = dataset.plot.scatter(x='ArrDelay', y='DepDelay', color='g')

plt.title('Retrasos')
plt.show()

#grfica 5: Tres variable(ArrDelay, DepDelay i ArrTime)
boxplot = dataset.boxplot(column=['ArrDelay','DepDelay','ArrTime'])

```
