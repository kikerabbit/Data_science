```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
plt.figure(figsize=(16,9))
plt.axis("off")
img = mpimg.imread('arale.png')
plt.imshow(img[:,:,0],vmin=0,vmax=1)
plt.title("Canal Vermell")
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img[:,:,1],vmin=0,vmax=1)
plt.title("Canal Verd")
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img[:,:,2],vmin=0,vmax=1)
plt.title("Canal Blau")

img_red=np.copy(img)
img_red[:,:,1]=0
img_red[:,:,2]=0
plt.title("Arale Vermella")
plt.imshow(img_red)
plt.savefig("Arale_vermella.png")
plt.show()
```
